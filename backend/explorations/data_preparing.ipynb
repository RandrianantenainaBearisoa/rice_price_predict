```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
fuel_price_file = 'data/data_lake/global_fuel_price.xlsx'
food_price_file = 'data/data_lake/mdg_food_price.csv'
```

```python
gasoline_data = pd.read_excel(fuel_price_file, sheet_name="Regular Gasoline (below RON 95)")
gasoline_price = gasoline_data[gasoline_data['Country'] == 'Madagascar']
```

```python
diesel_data = pd.read_excel(fuel_price_file, sheet_name="Diesel")
diesel_price = diesel_data[diesel_data["Diesel (LCU/liter)"] == "Madagascar"]
```

```python
kerosene_data = pd.read_excel(fuel_price_file, sheet_name="Kerosene")
kerosene_price = kerosene_data[kerosene_data["Kerosene (LCU/liter)"] == "Madagascar"]
```

```python
fuel_price = pd.DataFrame()
keys_date = [date for date in gasoline_price.keys() if isinstance(date, datetime)]
fuel_price["date"] = keys_date
```

```python
fuel_price["gasoline_price"] = gasoline_price[keys_date].values.flatten().astype(float)
fuel_price["diesel_price"] = diesel_price[keys_date].values.flatten().astype(float)
fuel_price["kerosene_price"] = kerosene_price[keys_date].values.flatten().astype(float)
fuel_price = fuel_price.ffill().bfill()
fuel_price
```

```python
first_record_date = fuel_price.iloc[0]["date"]
last_record_date = fuel_price.iloc[-1]["date"]
last_record_date, first_record_date
```

```python
rice_price = pd.read_csv(food_price_file)
rice_price = rice_price[rice_price["commodity"].str.contains("Rice")]
rice_price = rice_price[((rice_price["date"] >= str(first_record_date)) & (rice_price["date"] <= str(last_record_date)))]
```

```python
# fuel_price.to_csv("data/data_warehouse/prepared_fuel_price.csv")
# rice_price.to_csv("data/data_warehouse/prepared_rice_price.csv")
```

```python
rice_price["date"] = pd.to_datetime(rice_price["date"])
```

```python
rice_and_fuel_price = pd.merge_asof(
    rice_price.sort_values('date'), 
    fuel_price.sort_values('date'), 
    on='date', 
    direction='backward'
)
rice_and_fuel_price
```
